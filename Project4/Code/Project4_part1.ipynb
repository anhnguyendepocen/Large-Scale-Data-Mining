```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data/network_backup_dataset.csv')
sizeSum=df.groupby(['weekNo','dayOfWeek','workFlowId'],sort=False)#['workFlowId','size'].sum()
wf0=sizeSum['size'].sum()[:,:,'work_flow_0']
wf1=sizeSum['size'].sum()[:,:,'work_flow_1']
wf2=sizeSum['size'].sum()[:,:,'work_flow_2']
wf3=sizeSum['size'].sum()[:,:,'work_flow_3'
wf4=sizeSum['size'].sum()[:,:,'work_flow_4']
```

```python
#Question 1 a
plt.close()
dayName=['Monday1','Tuesday1','Wednesday1','Thursday1','Friday1','Saturday1','Sunday1','Monday2','Tuesday2','Wednesday2','Thursday2','Friday2','Saturday2','Sunday2','Monday3','Tuesday3','Wednesday3','Thursday3','Friday3','Saturday3']

plt.plot(range(1,21),np.array(wf0)[0:20],'r')
plt.plot(range(1,21),np.array(wf1)[0:20],'g')
plt.plot(range(1,21),np.array(wf2)[0:20],'b')
plt.plot(range(1,21),np.array(wf3)[0:20],color='orange')
plt.plot(range(1,21),np.array(wf4)[0:20],color='gray')
plt.legend(['wf0','wf1','wf2','wf3','wf4'])
plt.xticks(range(1,21),dayName, rotation='vertical')
plt.show()
```
